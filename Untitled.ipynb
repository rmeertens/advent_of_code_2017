```python
class double_linked_list:
    def __init__(self,value, left=None, right=None):
        self.value = value
        if not left:
            self.next = self
            self.prev = self
        else:
            self.next = right
            self.prev = left
            
    def insert(self, value):
        newnode = double_linked_list(value, left=self, right=self.next)
        self.next.prev = newnode
        self.next = newnode
        return newnode
    def nextt(self):
        return self.next
    

currentnode = double_linked_list(0)
position = 0
puzzleinput = 377

for iteration in range(1,50000000+1):
    for p in range(puzzleinput):
        currentnode = currentnode.nextt()
    currentnode = currentnode.insert(iteration)
    
print("done: " + str(currentnode.next.value))
```
